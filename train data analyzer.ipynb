```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df =pd.read_csv('/content/loan_data.csv')
df.head()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['person_income'], bins=30, kde=True)
plt.title("Applicant Income Distribution")
plt.show()
```

```python
pd.crosstab(df['credit_score'], df['loan_status'])

sns.countplot(x='credit_score', hue='loan_status', data=df)
plt.title("credit_score vs loan_status")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['loan_status'], bins=20, kde=True)
plt.title("loan_status")
plt.show()
```

```python
print("loan_percent_income:", df['loan_percent_income'].mean())
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/ecommerce_product_dataset.csv')
df.head()
```

```python
df.info()
df.describe()
```

```python
sns.histplot(df['Price'], bins=20)
plt.title("Discount")
plt.show()
```

```python
df['Category'].value_counts().plot(kind='bar')
plt.title("Most Preferred Products")
plt.show()
```

```python
df['ProductID'].value_counts().head(10).plot(kind='bar')
plt.title("Top Customers by Purchase Frequency")
plt.show()
```

```python
print("Total Customers:", df['ProductID'].nunique())
print("Sales:", df['Sales'].mean())
print("Most Popular Product:")
print(df['ProductID'].mode()[0])
```
