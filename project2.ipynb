```python
batsman = pd.read_csv('/content/sharma-kohli copy.csv')
batsman.head()
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('V Kohli Vs RG Sharma')
plt.xlabel('Year')
plt.ylabel('runs scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color = '#D9F10F')
plt.plot(batsman['index'],batsman['RG Sharma'],color = '#FC00D6')
plt.title('V Kohli Vs RG Sharma')
plt.xlabel('Year')
plt.ylabel('runs scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color = '#D9F10F',linestyle='solid',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color = '#FC00D6',linestyle='dashed',linewidth=2)
plt.title('V Kohli Vs RG Sharma')
plt.xlabel('Year')
plt.ylabel('runs scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color = '#D9F10F',linestyle='solid',linewidth=3,marker = 'D',markersize = 10)
plt.plot(batsman['index'],batsman['RG Sharma'],color = '#FC00D6',linestyle='dashed' ,linewidth=2,marker ='o')
plt.title('V Kohli Vs RG Sharma')
plt.xlabel('Year')
plt.ylabel('runs scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color = '#D9F10F',linestyle='solid',linewidth=3,marker = 'D',markersize = 10,label ='Virat Kohli')
plt.plot(batsman['index'],batsman['RG Sharma'],color = '#FC00D6',linestyle='dashed' ,linewidth=2,marker ='o',label='RG Sharma')
plt.title('V Kohli Vs RG Sharma')
plt.xlabel('Year')
plt.ylabel('runs scored')
```
